Из A в B с интервалом 10 мин. идут с постоянной скоростью, равной 60 км/ч, два поезда. С какой постоянной
скоростью (км/ч) идёт встречный поезд, если он повстречал эти поезда через 6 мин. один после другого? 

```python
v1 = 60; t1 = 10/60; t2 = 6 / 60
s = v1 * t1
vr = s / t2
v2 = vr - v1
print(v2,"km/h")
```

```python
from sympy.abc import x
from sympy import solve
from sympy import Eq

sol = solve(Eq(6 * (60 + x) / 60, 60 * 10 / 60), x)
print(sol,"km/h")
```
